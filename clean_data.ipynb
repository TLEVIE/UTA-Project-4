```python
## Used Kaggle Data @ https://www.kaggle.com/datasets/neisse/scrapped-lyrics-from-6-genres?resource=download
```

```python
import pandas as pd
```

```python
artists = pd.read_csv('artists-data.csv')
```

```python
artists.shape
```

```python
artists.head()
```

```python
lyrics = pd.read_csv('lyrics-data.csv')
lyrics.shape
```

```python
lyrics.head()
```

```python
data = lyrics.join(artists.set_index('Link'), on='ALink')
data = data.dropna()
data
```

```python
data['language'].value_counts()
```

```python
data = data.loc[data['language'] == 'en']
data['language'].value_counts()
```

```python
data.head()
```

```python
data.drop_duplicates(subset=['SName'], inplace=True)
```

```python
data = data.join(data['Genres'].str.split(pat=';',expand=True).add_prefix('Genres'))
data.head()
```

```python
data['Genres0'].value_counts()
```

```python
data.drop(data.columns.difference(['SName','Lyric','Artist','Popularity','Genres0']),1, inplace=True)
```

```python
data.dropna(inplace=True)
data.shape
```

```python
data.head()
```

```python
data_b = data.rename(columns={'SName': 'Song_Name', 'Genres0':'Genre'})
```

```python
data_b['Lyric'] =data_b['Lyric'].str.replace(',','')
data_b
```

```python
from sklearn.model_selection import train_test_split
train, val = train_test_split(data_b, test_size=0.1)
```

```python
train.to_csv('train.csv', header=True, index=False)
val.to_csv('val.csv', header=True, index=False)
```
